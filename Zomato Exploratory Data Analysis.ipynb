```python
df_rating = df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
df_rating
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Counts',hue='Rating color',data=df_rating,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x='Rating color',data=df_rating,palette=['blue','red','orange','yellow','green','green'])
```
